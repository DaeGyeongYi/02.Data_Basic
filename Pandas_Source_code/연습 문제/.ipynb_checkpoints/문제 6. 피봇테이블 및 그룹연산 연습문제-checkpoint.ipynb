#### 연습문제 1
- 붓꽃(iris) 데이터에서 붓꽃 종(species)별로 꽃잎길이(sepal_length), 꽃잎폭(sepal_width) 등의 평균을 구하라

```python
# 붓꽃(iris) 데이터 로드
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
# 붓꽃 종 별로 그룹핑
groups = iris.groupby('species')
pd.DataFrame(groups)
pd.DataFrame(groups)[1][0]


```

```python
# 그룹핑된 위 객체를 이용해 각 품종 별 꽃잎길이의 평균을 구하시오
groups.mean().petal_length
```

```python
# 그룹핑된 groups 객체를 이용하여 각 품종별 꽃받침 길이의 평균을 구하시오
groups.mean().sepal_length
```

#### 문제 1

- total_bill 대비 팁 요금의 비율을 계산하고 가공필드 tip_pt 열에 저장하시오.

- 팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받는지 살펴본다.

- 어떤 요인이 가장 크게 작용하는지 판단할 수 있는 방법이 있는가?

```python
tips['tip_pt'] = tips['tip']/tips['total_bill']
```

```python
# total_bill 대비 팁 요금의 비율을 계산하고 가공필드 tip_pt 열에 저장하시오.

tips.head()
tips.tail()
```

```python
# 각 요일별 손님의 수를 확인하세요
tips[['day']].value_counts().sort_index()
```

```python
# 각 요일별 한테이블당 인원수의 평균을 확인하세요
tips.loc[:,['day','size']].groupby('day').mean()['size'].round(4)
```

```python
# 점심/저녁 손님의 수를 비교 하세요
tips['time'].value_counts().sort_values()
```

```python
# 점심 저녁 별로 테이블당 인원수의 평균을 구하시오.
tips.groupby('time').mean()[['size']]
```

```python
# 각 요일별 점심/저녁 손님의 수를 피봇 테이블로 확인하세요
pd.pivot_table(tips,
              index='day',
              columns = 'time',
              values='size',
               aggfunc='count')
```

```python
# 각 요일별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
pd.pivot_table(tips,
              index=['day','time'],
              values='size',
               aggfunc=['mean','max','min']).fillna(0).round(2)
```

```python
# 각 요일별 테이블당 인원수의 평균과 최대인원 최소인원을 피봇테이블로 확인하시오
pd.pivot_table(tips,
              index='time',
               columns='day',
              values='size',
               aggfunc=['mean','max','min']).round(2).fillna(0)
```

```python
# 각 요일별 tip_pt의 평균을 구하시오
pd.pivot_table(tips,
              index='day',
              values='tip_pt')
```

```python
# 시간대별 tip_pt의 평균을 확인하시오
pd.pivot_table(tips,
              index='time',
              values='tip_pt')
```

```python
# 각 요일에 따른 time별 tip_pt의 평균을 확인하시오
pd.pivot_table(tips,
              index=['day','time'],
              values='tip_pt',
              fill_value = 0,
              aggfunc = ['mean']
              )
```

```python
# 각 요일별로 시간대 평균인원수와 팁비율의 평균을 확인하시오

pd.pivot_table(tips,
              index=['day','time'],
              values=['tip_pt','size'],
              fill_value = 0,
              aggfunc = 'mean'
              ).round(4)
```

#### 연습문제 3

- 타이타닉 승객 데이터를 이용하여 다음 분석을 실시하라. 데이터는 다음과 같이 받을 수 있다.
    - titanic = sns.load_dataset("titanic")
    
- 문제 1

- qcut 명령으로 세 개의 나이 그룹['청년','중년','장년']을 만든다.



- 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 
    - 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다. 
    - 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.
    

- 성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
# 데이터 로드하기
import seaborn as sns
titanic = sns.load_dataset("titanic")
```

```python
# qcut 명령으로 세 개의 나이 그룹['청년','중년','장년']의 범주형 값을 만들어  age_group 열을 생성한다
titanic['age_group'] = pd.qcut(titanic['age'],3,labels=['청년','중년','장년'])
```

```python
titanic.head()
```

- 성별, 선실, 나이 그룹에 의한 생존율을  계산한다. 

    - 행에는 성별로 나이그룹을 구분하는 다중 인덱스를 정의하고 열에는 class별 생존구분을 표현하는 pivot 테이블을 생성한다 

```python
# 피봇 테이블 생성 

titanic_df = pd.pivot_table(titanic,
              index=['sex','age_group'],
              columns=['class','alive'],
               values='survived',
              aggfunc=['count'])
```

```python
titanic_df
```

```python
# First 선실의 성별로 age_group의 인원 계산 코드
# First 선실의 청년 중년 장년의 각 인원수는 22,19,44 명임
titanic_df[('count','First')].sum(axis=1)
```

```python
data= {'First': titanic_df
       'Second':#Second 생존율 계산,
       'Third' : #Third 생존율 계산
      }

titanic_survived_rate = pd.DataFrame(data,columns=['First','Second','Third'])
titanic_survived_rate
```

```python
# 생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.
# 생존율 계산 후 데이터 프레임 생성
data= {'First': titanic_df[('count','First', 'yes')]/titanic_df[('count','First')].sum(axis=1),
       'Second':titanic_df[('count','Second', 'yes')]/titanic_df[('count','Second')].sum(axis=1),
       'Third' : titanic_df[('count','Third', 'yes')]/titanic_df[('count','Third')].sum(axis=1)
      }

titanic_survived_rate = pd.DataFrame(data,columns=['First','Second','Third'])
titanic_survived_rate
```

```python
# titanic_survived_rate 데이터프레임을 사용해 성별 각 선실의 생존율을 피봇 데이로 생성
pd.pivot_table(titanic_survived_rate,
              index = 'sex').round(4)*100
```

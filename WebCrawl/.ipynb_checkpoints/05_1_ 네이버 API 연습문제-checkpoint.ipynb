## url 
- 데이터 형태에 따라 2개의 url을 제공
- https://openapi.naver.com/v1/search/blog.xml : xml data 반환 
- https://openapi.naver.com/v1/search/blog.json (.json은 생략가능) : json 반환



```python
url_base = 'https://openapi.naver.com/v1/search/blog?query='  #query=검색어
```

```python
url = url_base + enc_text
url
```

#### header 부착(id,key 전송)

- urllib.request.Request(url): 전송객체 생성
- Request.add_headr(헤더) : 헤더 부착
- request.urlopen(request 객체): 요청

```python
request = urllib.request.Request(url) #요청한 객체

#헤더 추가
request.add_header('X-Naver-Client-Id',client_id)
request.add_header('X-Naver-Client-Secret',client_secret)

#요청 후 응답
response = urllib.request.urlopen(request)

#서버가 전송한 응답코드 확인
rescode = response.getcode()
rescode
```

```python
if(rescode==200):
    response_body = response.read()  #전송된 검색 data 읽어오기
  #  print(response_body) #encoding된 결과가 넘어옴
    print(response_body.decode('utf-8')) #한글 처리를 해야 볼 수 있다.
else:
    print("Error Code:" + rescode)
```

```python
import requests #서버접속 [위의 코드를 조금 간단하게 변경]
from urllib.parse import urlparse #한글처리
import urllib.parse

keyword = urllib.parse.quote("강남역")
url ="https://openapi.naver.com/v1/search/blog?query="
url_fin = url+keyword

#get()안에 url과 headers를 포함 할 수 있음
result = requests.get(url_fin,
      headers={"X-Naver-Client-Id":client_id,
               "X-Naver-Client-Secret":client_secret})

result.json()  #한글 처리와 data형식을 json으로 변경할 수 있음
```

```python
## 연습문제
## 원하는 검색어를 설정한 후 

## 글 제목, 설명, 세부링크, 블로거명, 블로그 주소, 생성날짜를 추출해서 df로 저장
## 200개 데이터를 저장해 볼 것
```

```python
import requests #서버접속 [위의 코드를 조금 간단하게 변경]
from urllib.parse import urlparse #한글처리
import urllib.parse

keyword = urllib.parse.quote("레드벨벳")
url ="https://openapi.naver.com/v1/search/blog?query="
url_fin = url+keyword+'&display=100'

#get()안에 url과 headers를 포함 할 수 있음
result = requests.get(url_fin,
      headers={"X-Naver-Client-Id":client_id,
               "X-Naver-Client-Secret":client_secret})

result.json()  #한글 처리와 data형식을 json으로 변경할 수 있음
```

```python
json_obj = result.json()
print(json_obj["lastBuildDate"])
print(json_obj['total'])
print(json_obj['start'])
print(json_obj['display'])
```

```python
#글 제목, 설명, 세부링크, 블로거명, 블로그 주소, 생성날짜를 추출해서 df로 저장
title_list = []
blogger_name = []
blogger_link = []
description = []
link = []
date = []



for item in json_obj['items']:
    title_list.append(item['title'].replace('<b>',"").replace("</b>",""))
    link.append(item['link'])
    description.append(item['description'].replace('<b>',"").replace("</b>",""))
    blogger_name.append(item['bloggername'])
    blogger_link.append(item['bloggerlink'])
    date.append(item['postdate'])
```

```python
import pandas as pd
df= pd.DataFrame({"글제목":title_list,
                  "설명":description,
                  "세부링크":link,
                  "블로거명":blogger_name,
                  "블로그주소":blogger_link,
                  "생성날짜":date})
```

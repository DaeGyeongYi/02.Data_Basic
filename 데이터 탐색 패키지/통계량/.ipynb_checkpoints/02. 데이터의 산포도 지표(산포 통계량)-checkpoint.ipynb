### ë¶„ì‚°(variance)

- ì‚°í¬ë„ì˜ ì²™ë„ë¡œ ê°€ì¥ ë„ë¦¬ ì‚¬ìš©ë˜ëŠ” ë°©ë²•
- ë°ì´í„°ê°€ í¼ì ¸ìˆëŠ” ì •ë„ì˜ ê¸°ì¤€ìœ¼ë¡œ í‰ê· ì„ ì‚¬ìš©

- ê³„ì‚° ë°©ë²•
    - ê° ë°ì´í„°ì™€ í‰ê· ì˜ ì°¨ì´ë¥¼ ì œê³±í•˜ì—¬ í•©í•œ ê°’ì˜ í‰ê· 
![](ë¶„ì‚°ì‹.png)

    - ë°ì´í„°ê°€ ëª¨ì§‘ë‹¨ ì „ì²´ì¼ ê²½ìš°ì—ëŠ” ë°ì´í„°ì˜ ê°œìˆ˜(n)ìœ¼ë¡œ ë‚˜ëˆ„ì–´ì¤Œ
    - í‘œë³¸ì¼ ê²½ìš° (n-1)ë¡œ ë‚˜ëˆ„ì–´ ì¤Œ
    - í‘œë³¸ì˜ ê²½ìš° nìœ¼ë¡œ ë‚˜ëˆ„ì–´ ì£¼ëŠ” ê²ƒë³´ë‹¤ (n-1)ë¡œ ë‚˜ëˆ„ì–´ ì£¼ëŠ” ê²ƒì´ ë” ì¢‹ì€ ì²™ë„ê°€ ë˜ê¸° ë•Œë¬¸ì¸ë° í‘œë³¸ì˜ í¬ê¸°ê°€ í° ê²½ìš°ì—ëŠ” ë³„ ì°¨ì´ê°€ ì—†ìŒ


- ë¶„ì‚°ê³„ì‚° : var(data, ddof=0|1) í•¨ìˆ˜ ì‚¬ìš© :ddof ìƒëµí•˜ë©´ 0

```python
## ddofì¸ìˆ˜ :  (ììœ ë„-ëª¨ìˆ˜ì§‘ë‹¨ëƒ í‘œë³¸ì´ëƒ)ëŠ” ê°’ì„ 1ë¡œ ë‘ê³  ì‚¬ìš©í•œë‹¤ê³  ìƒê°í•˜ë©´ í¸í•¨
## íŠ¹ë³„í•œ ê²½ìš° ì œì™¸í•˜ê³ ëŠ” ëª¨ë‘ sample ë°ì´í„°ì´ë¯€ë¡œ ë¶„ëª¨ë¥¼ n-1ë¡œ ë‘ 
## ì¦‰ , ddofëŠ” 1ë¡œ ë‘”ë‹¤ ë³´í†µì€ í‘œë³¸ì„ê°€ì§€ê³  í•˜ë‹ˆê¹Œ.

x =[1,2,3,4,5]

np.var(x, ddof=1) #í‘œë³¸ì„ ê¸°ì¤€ìœ¼ë¡œ ë¶„ëª¨ n-1ë¡œ ë‚˜ëˆ„ê² ë‹¤
np.var(x) #ëª¨ì§‘ë‹¨ ì „ì²´ë¥¼ ëŒ€ìƒìœ¼ë¡œ ì‚¬ìš©, ë¶„ëª¨ nìœ¼ë¡œ ë‚˜ëˆ„ê² ë‹¤
np.array(x).var()
pd.Series(x).var(ddof=0)

#ê°’ì˜ ìŠ¤ì¼€ì¼ì— í¬ê²Œ ì˜í–¥ì„ ë°›ìœ¼ë¯€ë¡œ, ë³€ìˆ˜ë¥¼ ìŠ¤ì¼€ì¼ë§í•œ í›„ ë¶„ì‚°ê³„ì‚°
# í˜¹ì€ í‘œì¤€í¸ì°¨ë¥¼ í™œìš©
```

### í‘œì¤€í¸ì°¨(standard deviation)
- ê³„ì‚°ëœ ë¶„ì‚°ì˜ ì œê³±ê·¼ìœ¼ë¡œ ê³„ì‚°
- í‰ê· ì„ ì¤‘ì‹¬ìœ¼ë¡œ ì¼ì •í•œ ê±°ë¦¬ì— í¬í•¨ëœ ë°ì´í„°ì˜ ë¹„ìœ¨ì´ ì–¼ë§ˆì¸ê°€ë¥¼ ê²Œì‚°
- ëª¨ë“  ë°ì´í„°ë¥¼ ê³ ë ¤í•œ ì²™ë„

- íŠ¹ì§•
    - ëª¨ë“  ë°ì´í„°ê°€ ë™ì¼í•œ ê°’ì„ ê°–ëŠ”ë‹¤ë©´ ë¶„ì‚°ê³¼ í‘œì¤€í¸ì°¨ëŠ” 0ìœ¼ë¡œ ê³„ì‚°
    - ëª¨ë“  ë°ì´í„°ì— ë™ì¼í•œ ê°’ì„ ë”í•´ ì£¼ê±°ë‚˜ ë¹¼ë„ ë³€í•˜ì§€ ì•ŠìŒ
    - ëª¨ë“  ë°ì´í„°ì— ë™ì¼í•œ ê°’(C)ì„ ê³±í•˜ë©´ ë¶„ì‚°ì€ $ë¶„ì‚°Ã—C^2$ìœ¼ë¡œ í‘œì¤€í¸ì°¨ëŠ” $í‘œì¤€í¸ì°¨Ã—C$ ë§Œí¼ ì»¤ì§

#### í‘œì¤€ í¸ì°¨ ê³„ì‚°: std() í•¨ìˆ˜ ì‚¬ìš©

- ë¶„ì‚°ì—ì„œ ì œê³±ì˜ ì˜í–¥ì„ ì—†ì•¤ ì§€í‘œ
- ë¶„ì‚°ê³¼ í‘œì¤€í¸ì°¨ê°€ í¬ë©´ í´ìˆ˜ë¡ ì‚°í¬ë„ê°€ ë†’ë‹¤

```python
x = [1,2,3,4,5]
np.std(x, ddof = 1)
np.array(x).std(ddof =0)
pd.Series(x).std(ddof=1)
```

#### ë³€ë™ê³„ìˆ˜(CV: Coefficient of Variable)

- í‘œë³¸ í‘œì¤€ í¸ì°¨ë¥¼ í‘œë³¸ í‰ê· ìœ¼ë¡œ ë‚˜ëˆˆ ê°’ ë˜ëŠ” ê·¸ ê°’ì— 100ì„ ê³±í•œê°’

- ì„œë¡œ ë‹¤ë¥¸ í‰ê· ê³¼ í‘œì¤€ í¸ì°¨ë¥¼ ê°–ëŠ” ì—¬ëŸ¬ ë°ì´í„°ì˜ í©ì–´ì§„ ì •ë„ë¥¼ ë¹„êµí•  ë•Œ ì‚¬ìš©

- ë³€ë™ê³„ìˆ˜ ê°’ì´ í¬ë‹¤ëŠ” ê²ƒì€ ë°ì´í„°ê°€ ì‚°í¬ëœ ì •ë„ê°€ ìƒëŒ€ì ìœ¼ë¡œ í¬ë‹¤ëŠ” ì˜ë¯¸
- í‘œë³¸ ë³€ë™ê³„ìˆ˜ $ ğ¶ğ‘‰=\frac{S}{\overline{x}}$,  ëª¨ë³€ë™ê³„ìˆ˜  $ğ¶ğ‘‰= \frac{ğœ}{ğœ‡}$

#### ë³€ë™ê³„ìˆ˜ ê³„ì‚° í•¨ìˆ˜: variation(data)
- ëª¨ë“  ê³„ìˆ˜ê°€ ì–‘ìˆ˜ê°€ ì•„ë‹ˆë©´ ì˜ ì“°ì§€ ì•ŠìŒ
- ììœ ë„ëŠ” 0ìœ¼ë¡œ ì…‹íŒ… ë˜ì–´ìˆìŒ

#### ë³€ë™ê³„ìˆ˜ì˜ í•„ìš”ì„±
- í‘œì¤€í¸ì°¨ë¥¼ êµ¬í• ë•Œ ìŠ¤ì¼€ì¼ì— ì˜í–¥ì„ ë°›ì§€ ì•Šê¸° ìœ„í•´ì„œ ë³€ìˆ˜ ìŠ¤ì¼€ì¼ë§ì„ í•œ í›„ í¸ì°¨ë¥¼ êµ¬í•¨
    - ë°ì´í„°ê°€ ë³´ë‘ ì–‘ìˆ˜ì¸ ê²½ìš°ì—ëŠ” ë³€ë™ê³„ìˆ˜ë¥¼ ì‚¬ìš©í•¨
    

```python

x1 = np.array([1,2,3,4,5])
x2 = x1*10
x2
```

```python
np.std(x1, ddof=1)
np.std(x2, ddof=1) #í‘œì¤€í¸ì°¨ëŠ” *10ë§Œí¼ì˜ ì°¨ì´ê°€ ë‚ ê²ƒ (x1ì— 10ì”© ê³±í•´ì§„ê±°ë‹ˆê¹Œ)
```

```python
variation(x1)
variation(x1)
```

```python
# ë³€ë™ê³„ìˆ˜ ìˆ˜ì‹
np.std(x1, ddof=1)/np.mean(x1)
np.std(x2, ddof=1)/np.mean(x2)
```

#### ìŠ¤ì¼€ì¼ë§(í‘œì¤€í™”)

- í‰ê· ì€ 0ìœ¼ë¡œ í‘œì¤€í¸ì°¨ëŠ” 1ì´ë¨
- ê° ê°’ë“¤ì„ ìƒëŒ€ì ì¸ ê°’ìœ¼ë¡œ ë³€í™”ì‹œí‚¤ëŠ” ê¸°ë²•
- í‰ê· ì´ 95ì (êµ­ì–´), 30ì (ìˆ˜í•™) ì¸ ë°˜ì—ì„œ
    - ê¸¸ë™ì´ê°€ ë°›ì€ ì ìˆ˜ 90(êµ­ì–´), 80ì (ìˆ˜í•™) ì´ë¼ë©´ ê¸¸ë™ì´ëŠ” ì–´ë–¤ ê³¼ëª©ì„ ë” ì˜í•œ ê²ƒì¼ê¹Œ?
    
- ìƒëŒ€ì ìœ¼ë¡œ ë‹¨ìœ„ë¥¼ ì¼ì •í•˜ê²Œ ë§ì¶°ì„œ ë¹„êµí•  ìˆ˜ ìˆê²Œí•´ì¤€ë‹¤

- í‘œì¤€í™”ëœ ë°ì´í„°ëŠ” ì ìˆ˜ì™€ ê°™ì€ ë‹¨ìœ„ë¥¼ ì‚¬ìš©í•˜ì§€ ì•ŠìŒ

```python
x1
x2
```

- standard scaling - í‰ê· ì´ 0, í‘œì¤€í¸ì°¨ê°€ 1ìœ¼ë¡œ ë°”ê¾¸ëŠ” í‘œì¤€í™”
- (ë³€ëŸ‰ - í‰ê· ) / í‘œì¤€í¸ì°¨

```python
z1 = (x1 - x1.mean())/x1.std()

z2 = (x2 - x2.mean())/x2.std()

z1
z2

#ë³´í†µ -3 ~ +3 ê¹Œì§€ ë¶„í¬ë¨
```

```python
z1.mean()
z2.mean()
z1.std(ddof=0)

#í‰ê· ì€ 0ì´ ë˜ì—ˆê³ , í‘œì¤€í¸ì°¨ëŠ” 1ì´ ë˜ì—ˆìŠµë‹ˆë‹¹
```

#### MinMax ìŠ¤ì¼€ì¼ë§

```python
z1 = (x1 - x1.min()) / (x1.max() - x1.min())
z2 = (x2 - x2.min()) / (x2.max() - x2.min())

x1
x2

print(z1)
print(z2)
```

- í•™ë ¥ ë“±ì˜ ê²€ì‚¬ ê²°ê³¼ê°€ ì§‘ë‹¨ì˜ í‰ê· ê°’ì—ì„œ ì–´ëŠ ì •ë„ ë–¨ì–´ì ¸ ìˆëŠ”ê°€ë¥¼ ìˆ˜ì¹˜ë¡œ ë‚˜íƒ€ë‚¸ ê²ƒ.
